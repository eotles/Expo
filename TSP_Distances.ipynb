```python
import pandas as pd
rel_fp = 'WI_TSP_Distances.csv'
```

```python
data = pd.read_csv(rel_fp)
data = data.sort('City')
data = data.to_dict(orient='records')
```

```python
data
```

```python
def distance(row1, row2):
    distance = 0
    for dim in ['X', 'Y']:
        distance += (row1[dim] - row2[dim])**2
    return(distance**0.5)
    
distance_data = []
for row1_idx, row1 in enumerate(data):
    distance_data.append({'City 1': row1['City'], 'City 2': row1['City'], 'Distance': 0})
    for row2 in data[row1_idx+1:]:
        d = distance(row1, row2)
        distance_data.append({'City 1': row1['City'], 'City 2': row2['City'], 'Distance': d})
        distance_data.append({'City 1': row2['City'], 'City 2': row1['City'], 'Distance': d})
distance_data = pd.DataFrame(distance_data)
```
